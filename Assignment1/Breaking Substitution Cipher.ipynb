```python

#  Frequency analysis


from collections import Counter
import math

# remove line breaks for analysis
flat = clean_cipher.replace('\n', '')

# count all characters
char_counts = Counter(flat)

total_chars = sum(char_counts.values())
print("Total characters:", total_chars)

print("\n=== Character frequencies (sorted) ===")
for ch, cnt in char_counts.most_common():
    freq = cnt / total_chars * 100
    # show space as '[space]' for readability
    printable = '[space]' if ch == ' ' else ch
    print(f"{printable!r}: {cnt} ({freq:.2f}%)")

# Bigram frequencies (pairs of chars)
bigrams = Counter(zip(flat, flat[1:]))
total_bigrams = sum(bigrams.values())

print("\n=== Top 50 bigrams ===")
for (a, b), cnt in bigrams.most_common(50):
    a_p = '[space]' if a == ' ' else a
    b_p = '[space]' if b == ' ' else b
    print(f"{a_p}{b_p}: {cnt}")
```

```python


import string
from collections import Counter

symbols_only = ''.join(
    ch for ch in flat
    if ch != ' ' and ch not in string.digits and ch not in string.whitespace
)

sym_counts = Counter(symbols_only)
total_sym = sum(sym_counts.values())

print("Total non‑space, non‑digit symbols:", total_sym)
print("\n=== Symbol frequencies (sorted) ===")
for ch, cnt in sym_counts.most_common():
    freq = cnt / total_sym * 100
    print(f"{ch!r}: {cnt} ({freq:.2f}%)")

# we compare it to english letter frequencies


# Approximate English letter frequency order (most -> least)
english_letters_by_freq = list("etaoinshrdlcumwfgypbvkjxqz")

print("\n=== English letters by typical frequency ===")
print(" ".join(english_letters_by_freq))

# Get symbols sorted by frequency (cipher side)
cipher_symbols_by_freq = [ch for ch, _ in sym_counts.most_common()]

print("\n=== Cipher symbols by frequency ===")
print(" ".join(cipher_symbols_by_freq))

#  rank-based mapping : cipher[i] -> english[i]
rank_mapping = {}
for i, sym in enumerate(cipher_symbols_by_freq):
    if i < len(english_letters_by_freq):
        rank_mapping[sym] = english_letters_by_freq[i]
    else:
        break

print("\n=== Naive rank-based suggestion (cipher -> plaintext) ===")
for sym, plain in rank_mapping.items():
    print(f"{sym!r} -> {plain}")


# Bigram (2-symbol) analysis


def ngrams(seq, n):
    return [seq[i:i+n] for i in range(len(seq) - n + 1)]

bigrams = ngrams(symbols_only, 2)
bigram_counts = Counter(bigrams)
total_bigrams = sum(bigram_counts.values())

print("\n=== Top 50 cipher bigrams ===")
for bg, cnt in bigram_counts.most_common(50):
    freq = cnt / total_bigrams * 100
    print(f"{bg!r}: {cnt} ({freq:.2f}%)")

# Common English bigrams (reference)
english_bigrams = [
    "th","he","in","er","an","re","on","at","en","nd",
    "ti","es","or","te","of","ed","is","it","al","ar"
]

print("\n=== Common English bigrams (reference) ===")
print(" ".join(english_bigrams))

# Trigram (3-symbol) analysis

trigrams = ngrams(symbols_only, 3)
trigram_counts = Counter(trigrams)
total_trigrams = sum(trigram_counts.values())

print("\n=== Top 50 cipher trigrams ===")
for tg, cnt in trigram_counts.most_common(50):
    freq = cnt / total_trigrams * 100
    print(f"{tg!r}: {cnt} ({freq:.3f}%)")

# Common English trigrams
english_trigrams = [
    "the","and","ing","her","hat","his","tha","ere","for","ent",
    "ion","ter","was","you","ith","ver","all","wit","thi"
]

print("\n=== Common English trigrams  ===")
print(" ".join(english_trigrams))

# bigram rank alignment (cipher - english)

cipher_bigrams_by_freq = [bg for bg, _ in bigram_counts.most_common(20)]

print("\n=== Cipher bigrams by frequency ===")
print(" ".join(cipher_bigrams_by_freq))

print("\n=== Rank-based bigram pairing  ===")
for cb, eb in zip(cipher_bigrams_by_freq, english_bigrams):
    print(f"{cb!r}  ≈  {eb}")

```
